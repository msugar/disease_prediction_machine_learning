```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def plot_feature_correlation_heatmap(data_path):
    """
    Loads a dataset from the specified path, calculates the Pearson correlation 
    between features, and plots a heatmap of the correlations.

    Parameters:
    data_path (str): The path to the CSV dataset.

    Returns:
    None: Displays the correlation heatmap.
    """
    # Load the dataset
    disease_training = pd.read_csv(data_path)
    disease_training.drop(['Unnamed: 133'], axis=1, inplace=True)

    # Extract features
    X_train = disease_training.iloc[:, :-1]  # Exclude the last column (prognosis)

    # Calculate Pearson correlation between features
    feature_correlation = X_train.corr()

    # Plot the correlation matrix as a heatmap
    plt.figure(figsize=(12, 10))
    sns.heatmap(feature_correlation, annot=False, cmap='coolwarm', linewidths=0.5)
    plt.title('Feature Correlation Heatmap')
    plt.show()

# Usage:
plot_feature_correlation_heatmap("../../../data/raw/Training.csv")
```
